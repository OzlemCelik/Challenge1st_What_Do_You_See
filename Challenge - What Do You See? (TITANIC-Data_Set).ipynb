```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/Users/OzlemCelik/Desktop/all/train.csv")
```

```python
df.head(850)
```

```python
df.describe()
```

```python
df['survived'].hist(bins=50)
plt.ylabel('Number of people')
plt.xlabel('"0":Dead    "1":Survived')
plt.title('Number of Survived People')
#In dataset alive passengers represented as "1" and dead ones represented as "0"
#we can demonstrate number of alive and dead passengers by creating histogram plot
```

```python
df.boxplot(column='age')
plt.title('Age Distribution')

#We can use boxplot graph to show the distribution of passenger's age  
```

```python
plt.subplot(1,1,1)
plt.scatter(x = df['fare'], y = df['pclass'], color='green')
plt.ylabel('Class')
plt.xlabel('Fare')
plt.title('Compare Fare & Class')
plt.show()


#Scatter plot is one of the best way to compare two continuous data values.
#In that graph, Fare over class data is compared and it shows that the number of third class passengers 
#are larger than others.
```

```python
df.plot(kind='line', x='sex', y= 'survived')
plt.show()

#In this graph, Sex over Survived rate was going to be compared but I could retrieve any clear comparison.
```

```python
plt.subplot(1,1,1)
plt.scatter(x = df['age'], y = df['sex'], color='blue')
plt.ylabel('Sex')
plt.xlabel('Age')
plt.title('Compare Age & Sex Distirubition')
plt.show()

#In this graph, I just try to compare sex and age distirubition.
```
